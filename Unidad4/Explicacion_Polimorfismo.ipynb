```python
class Persona:
    def __init__(self, nombre, correo):
        self.nombre = nombre
        self.correo = correo

    def mostrar_info(self):
        return f"Nombre: {self.nombre}, Correo: {self.correo}"


class Estudiante(Persona):
    def __init__(self, nombre, correo, num_control, carrera):
        super().__init__(nombre, correo)
        self.num_control = num_control
        self.carrera = carrera

    def mostrar_info(self):
        return (f"[Estudiante] Nombre: {self.nombre}, "
                f"Número de Control: {self.num_control}, "
                f"Carrera: {self.carrera}")


class Trabajador(Persona):
    def __init__(self, nombre, correo, num_trabajador, departamento):
        super().__init__(nombre, correo)
        self.num_trabajador = num_trabajador
        self.departamento = departamento

    def mostrar_info(self):
        return (f"[Trabajador] Nombre: {self.nombre}, "
                f"Número de Trabajador: {self.num_trabajador}, "
                f"Departamento: {self.departamento}")
```

```python
personas = [
    Estudiante("María López", "maria@correo.com", "2023001", "Ingeniería"),
    Trabajador("Juan Pérez", "juan@correo.com", "T-001", "Biblioteca")
]

for p in personas:
    print(p.mostrar_info())
```
